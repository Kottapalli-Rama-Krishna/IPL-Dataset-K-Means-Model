```python
# Dealing with out-liers
for i in columns:
    if cricket[i].dtype!='object':
        plt.boxplot(cricket[i])
        plt.xlabel(i)
        plt.ylabel('count')
        plt.show()
```

```python
out_col_list = ['Not_Outs','Runs_Scored','Batting_Average','Balls_Faced','Batting_Strike_Rate','Centuries','Half_Centuries',
               'Fours','Sixes','Catches_Taken','Stumpings','Wickets_Taken',
                'Bowling_Average','Bowling_Strike_Rate','Four_Wicket_Hauls']
out_col_list
```

```python
# IQR Method
for i in out_col_list:
    Q1 = cricket[i].quantile(0.25)
    Q3 = cricket[i].quantile(0.75)
    IQR = Q3-Q1
    lower_range = Q1-1.5*IQR
    upper_range = Q3+1.5*IQR
    cricket_new = cricket[(cricket[i]>=lower_range)&(cricket[i]<=upper_range)]
```

```python
cricket_unbias = cricket.drop(["Player_Name"],axis=1)
cricket_unbias
```

```python
# Scaling to be done to standardize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
cricket_scaled = scaler.fit_transform(cricket_unbias)
cricket_scaled 
```

```python
cricket_scaled_dataset = pd.DataFrame(cricket_scaled, columns = ['Year', 'Matches_Batted', 'Not_Outs', 'Runs_Scored',
       'Highest_Score', 'Batting_Average', 'Balls_Faced',
       'Batting_Strike_Rate', 'Centuries', 'Half_Centuries', 'Fours', 'Sixes',
       'Catches_Taken', 'Stumpings', 'Matches_Bowled', 'Balls_Bowled',
       'Runs_Conceded', 'Wickets_Taken', 'Best_Bowling_Match',
       'Bowling_Average', 'Economy_Rate', 'Bowling_Strike_Rate',
       'Four_Wicket_Hauls', 'Five_Wicket_Hauls'])
cricket_scaled_dataset
```

```python
cricket_scaled_dataset['Best_Bowling_Match'].replace(np.nan,0,inplace=True)
```

```python
# Model building
from sklearn.cluster import KMeans
```

```python
# Elbow curve method to find the ideal number of clusters.
clusters=list(range(2,8))
ssd = []
for num_clusters in clusters:
    model_clus = KMeans(n_clusters = num_clusters, max_iter=150,random_state= 50)
    model_clus.fit(cricket_scaled_dataset)
    ssd.append(model_clus.inertia_)
```

```python
plt.plot(clusters,ssd)
```

```python
# we observed the sharp bend at 3
Kmodel = KMeans(n_clusters = 3, max_iter=150,random_state= 50)
Kmodel.fit(cricket_scaled_dataset)
```

```python
Kmodel.labels_
```

```python
cricket['Cluster_ID'] = Kmodel.labels_
```
